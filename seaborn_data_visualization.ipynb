```
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## NUMERICAL data point
1. relplot
2. scatterplot
3. lineplot


```
tips=sns.load_dataset('tips')
```

```
tips.head()
```

```
sns.relplot('total_bill','tip',data=tips)
```

```
sns.relplot('total_bill','tip',data=tips,hue='smoker')
```

```
sns.relplot('total_bill','tip',data=tips,hue='smoker',style='time')
```

```
sns.relplot('total_bill','tip',data=tips,hue='size')
```

```
sns.relplot('total_bill','tip',data=tips,size='size')
```

```
sns.relplot('total_bill','tip',data=tips,kind='line')
```

```
fmri=sns.load_dataset('fmri')
fmri.head()
```

```
sns.relplot(x='timepoint',y='signal',data=fmri)
```

```
sns.relplot(x='timepoint',y='signal',kind='line',data=fmri)
#at same time there is more values of signal
```

```
sns.relplot(x='timepoint',y='signal',kind='line',ci=False,data=fmri)
```

```
sns.relplot(x='timepoint',y='signal',kind='line',hue='event',data=fmri)
```

```
sns.relplot(x='timepoint',y='signal',kind='line',style='event',markers=True,data=fmri)
```

```
sns.relplot(x='timepoint',y='signal',hue='region',units='subject',estimator=None,kind='line',data=fmri.query("event=='stim'"))
```

```
dots=sns.load_dataset('dots').query("align=='dots'")
dots.head()
```

```
sns.relplot(x='time',y='firing_rate',data=dots)
```

```
sns.relplot(x='time',y='firing_rate',kind='line',data=dots)
```

```
sns.relplot(x='total_bill',y='tip',hue='smoker',col='time',data=tips)
```

```
sns.relplot(x='timepoint',y='signal',hue='subject',col='region',row='event',height=3,kind='line',estimator=None,data=fmri)
```

Categorical Data Plotting
1. catplot()
2. baxplot()
3. stripplot()
4. swarmplot()

```
titanic=sns.load_dataset('titanic')
titanic.head()
```

```
sns.catplot(x='day',y='total_bill',data=tips)
```

```
sns.catplot(y='day',x='total_bill',data=tips)
```

```
sns.catplot(x='day',y='total_bill',jitter=False,data=tips)
```

```
sns.catplot(x='day',y='total_bill',kind='swarm',data=tips)
```

```
sns.catplot(x='day',y='total_bill',hue='sex',data=tips)
```

```
sns.catplot(x='smoker',y='tip',order=['No','Yes'],data=tips)
```

```
sns.catplot(x='day',y='total_bill',hue='sex',kind='box',data=tips)
```

```
sns.catplot(x='day',y='total_bill',hue='sex',kind='boxen',data=tips)
```

```
sns.catplot(x='day',y='total_bill',hue='sex',kind='violin',data=tips)
```

```
g=sns.catplot(x='day',y='total_bill',kind='violin',data=tips)
sns.swarmplot(x='day',y='total_bill',color='k',size=3,ax=g.ax,data=tips)
```

```
sns.catplot(x='sex',y='survived',kind='bar',data=titanic)
#male servival chances are 0.2% and female servival chances are 0.75%. Black line represent uncertianity of prediction
```

```
sns.catplot(x='sex',y='survived',kind='bar',hue='class',data=titanic)
```

```
sns.catplot(x='sex',kind='count',data=titanic)
```

Visualize Distribution of data
1. distplot
2. kdeplot
3. jointplot
4. rugplot

```
x=np.random.randn(100)
```

```
sns.distplot(x)
```

```
sns.distplot(x,kde=False)
```

```
sns.distplot(x,kde=False,bins=30)
```

```
sns.distplot(x,hist=False,bins=30)
```

```
sns.distplot(x,hist=False,rug=True,bins=30)
```

```
sns.jointplot( x='total_bill',y='tip',data=tips)
```

```
sns.jointplot( x='total_bill',y='tip',kind='kde',data=tips)
```

Linear regression and Relation
1. regplot
2. lmplot

```
sns.regplot(x='total_bill',y='tip',data=tips)
```

```
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```
sns.lmplot(x='size',y='tip',data=tips)
```

```
data=sns.load_dataset('anscombe')
data.head()
```

```
sns.lmplot(x='x',y='y',data=data.query("dataset=='I'"),ci=None)
```

```
sns.lmplot(x='x',y='y',data=data.query("dataset=='II'"),ci=None)
#2nd order polynomial
```

```
sns.lmplot(x='x',y='y',data=data.query("dataset=='II'"),ci=None,order=2)
```

```
sns.lmplot(x='x',y='y',data=data.query("dataset=='III'"),ci=None)
#outlier
```

```
sns.lmplot(x='x',y='y',data=data.query("dataset=='III'"),ci=None,robust=True)
```
